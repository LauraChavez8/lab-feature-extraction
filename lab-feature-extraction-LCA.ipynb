```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
customer_df = pd.read_csv("/Users/laura.chavez/Documents/GitHub/lab-cleaning-numerical-data/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

Open the categoricals variable we created before.

```python
categoricals = customer_df.select_dtypes('object')
```

```python
categoricals.head()
```

Plot every categorical variable.

```python
for c in categoricals.columns:
    plt.figure(figsize=(10,4))
    plt.bar(categoricals[c].unique(),categoricals[c].value_counts())
    plt.title(c)
    plt.show()
```
